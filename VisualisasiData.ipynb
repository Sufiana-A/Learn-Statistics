```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('Solarize_Light2')
```

```python
file = "data-pelayanan-pengelolaan-sampah-b3-rumah-tangga-bulan-april-tahun-2021.csv"

df = pd.read_csv(file)

df.columns = [
    "Wilayah",
    "No Kecamatan",
    "Kemasan kaleng(kg)",
    "Kemasanplastik(kg)",
    "E-waste(kg)",
    "Baterai(kg)",
    "Lampu bekas(kg)",
    "Lb3 RT(kg)",
    "Limbah medis(kg)",
    "Total",
    "Oli bekas(L)"
]
```

```python
df = df.drop(columns=["No Kecamatan","Lb3 RT(kg)","Oli bekas(L)"])
x = df['Wilayah']
y = df['Total']
```

```python
#Plot the data using bar() method
plt.title("Data Penanganan Limbah DKI Jakarta April 2021")
plt.xlabel("Wilayah Jakarta", fontname='Comic Sans MS', fontsize=19)
plt.ylabel("Total Limbah (kg)", fontname='Comic Sans MS', fontsize=17)
plt.bar(x, y, color='g')
plt.show()
```

```python
plt.pie(y, labels=x, radius=1.4 ,autopct='%0.01f%%', shadow=True)
plt.show()
```
